Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

ANS-In Pandas, both loc and iloc are used to access data in a DataFrame, but they differ in the way they allow you to select data.

loc:
loc is primarily label-based indexing, meaning you use the actual row and column labels to access data.
When using loc, you specify the row and column labels as arguments.
The row and column labels can be strings, integers, or other data types assigned as index/column names.
It is inclusive of both the start and end values when slicing.
Example usage: df.loc[row_label, column_label] or df.loc[start_row_label:end_row_label, start_column_label:end_column_label]

iloc:
iloc is integer-based indexing, where you use integer-based positions to access data.
When using iloc, you provide the integer index positions to access data.
The row and column indices are represented by integers starting from 0.
It is exclusive of the end value when slicing, similar to how Python slicing works.
Example usage: df.iloc[row_index, column_index] or df.iloc[start_row_index:end_row_index, start_column_index:end_column_index]

Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

ANS- In pandas, a window function is used to perform calculations on a subset of data called a "window." The window is defined by a set of rows within a specified range. These functions are commonly used for calculating rolling statistics, aggregations, and other computations that require a window of data to be considered.

Types of window functions in pandas:

Rolling Windows: These functions operate on a fixed-size window and perform calculations like rolling mean, rolling sum, etc.

Expanding Windows: These functions consider all the rows up to the current row and perform cumulative calculations like cumulative sum, cumulative max, etc.

Exponentially Weighted Windows: These functions apply weights to the values within the window, giving more importance to recent data points.

Q8. Python program to calculate the difference between two dates:

ANS- import pandas as pd

def calculate_date_difference():
    # Prompting the user to enter dates
    date_str1 = input("Enter the first date (YYYY-MM-DD): ")
    date_str2 = input("Enter the second date (YYYY-MM-DD): ")

    # Converting the input strings to pandas datetime objects
    date1 = pd.to_datetime(date_str1)
    date2 = pd.to_datetime(date_str2)

    # Calculating the difference
    time_delta = date2 - date1

    # Extracting days, hours, and minutes from the time delta
    days_diff = time_delta.days
    hours_diff = time_delta.seconds // 3600
    minutes_diff = (time_delta.seconds % 3600) // 60

    # Displaying the result
    print("\nDifference:")
    print("Days:", days_diff)
    print("Hours:", hours_diff)
    print("Minutes:", minutes_diff)

calculate_date_difference()
